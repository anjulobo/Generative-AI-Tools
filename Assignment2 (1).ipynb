```python
#1. Create a Doc object from the file owlcreek.txt
```

```python
nlp = spacy.load("en_core_web_sm")

with open("owlcreek.txt", "r") as file:
    text = file.read()

doc = nlp(text)
```

```python
#2. How many tokens are contained in the file?
```

```python
num_tokens = len(doc)
print(f'Number of tokens in the file: {num_tokens}')
```

```python
#3. How many sentences are contained in the file?
```

```python
num_sentences = len(list(doc.sents))
print("Number of sentences in the file :", num_sentences)
```

```python
#4. Print the second sentence in the document
```

```python
sentences = list(doc.sents)
print("Second sentence:", sentences[1].text)
```

```python
#5. For each token in the sentence above, print its text, POS tag, dep tag and lemma.
```

```python
for sentence in sentences:
    for token in sentence:
        print(f"Text: {token.text}, POS: {token.pos_}, Dep: {token.dep_}, Lemmas: {token.lemma_}")
```

```python
#6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.
```

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)

pattern = [{'LOWER': 'swimming'}, {'LOWER': 'vigorously'}]

matcher.add('SWIMMING_VIGOROUSLY', [pattern])

matches = matcher(doc)

print(f"Found {len(matches)} occurrences of 'swimming vigorously':")
for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text)
```

```python
#7. Print the text surrounding each found match.
```

```python
if not matches:
    print("No matches were found, so there is no surrounding text to print.")
else:
    for match_id, start, end in matches:
           context_window = 10

    context_start = max(0, start - context_window)
    context_end = min(len(doc), end + context_window)

    full_span = doc[context_start:context_end]

    matched_span = doc[start:end]

    print(f"Match found: '{matched_span.text}'")
    print(f"Surrounding text: '{full_span.text}'")
    print("-" * 50)
```
