```python
#1 - Word Tokenization
#"Natural Language Processing is amazing!"

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Natural Language Processing is amazing!"
doc = nlp(text)

for token in doc:
    print(token.text)
```

```python
#2 - Sentence Tokenization
#"Python is easy. NLP is interesting."

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Python is easy. NLP is inresting."

doc = nlp(text)

sentences = [sent.text for sent in doc.sents]
print(sentences)
```

```python
#3 - Tokenization using split()
#"Machine Learning and AI"

text = "Machine Learning and AI"
tokens = text.split()
print(tokens)
```

```python
#5 - Remove Punctuation During Tokenization
#"Hello!!! How are you??"

import re

text = "Hello!!! How are you??"
tokens = re.findall(r'\b\w+\b', text)

print(tokens)
```

```python
#6 - Basic POS Tagging
#"Ravi plays cricket"

import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Ravi plays cricket")
for token in doc:
  print(token.text, token.pos_)


```

```python
#25 - Remove Stopwords
# "This is a sample sentence"

import spacy

nlp = spacy.load("en_core_web_sm")

text = "This is a sample sentence"

doc = nlp(text)

tokens = [token.text for token in doc if not token.is_stop]

print(tokens)
```

```python
#27 - Basic Phrase Matching
#"Machine learning is part of artificial intelligence"
#"machine learning", "artificial intelligence"

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")

text = "Machine learning is part of artificial intelligence"
doc = nlp(text)

matcher = PhraseMatcher(nlp.vocab)

phrases = ["Machine learning", "artificial intelligence"]
patterns = [nlp(phrase) for phrase in phrases]

matcher.add("TECH_TERMS", patterns)

matches = matcher(doc)

for match_id, start, end in matches:
    print(doc[start:end].text)
```

```python
#28 - Match Product Names
#"I bought an iPhone yesterday"
#"iPhone", "Samsung Galaxy"

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")

text = "I bought an iPhone yesterday"
doc = nlp(text)

matcher = PhraseMatcher(nlp.vocab)

product_names = ["iPhone", "Samsung Galaxy"]
patterns = [nlp(name) for name in product_names]

matcher.add("PRODUCT_NAMES", patterns)

matches = matcher(doc)

for match_id, start, end in matches:
    print(doc[start:end].text)
```

```python
#29 - Match City Names
#"Delhi is a busy city"
#"Delhi", "Mumbai"

import spacy
from spacy import matcher

nlp = spacy.load("en_core_web_sm")

text = "Delhi is a busy city"

cities = ["Delhi", "Mumbai"]

for city in cities:
    if city.lower() in text.lower():
        print("City Found:", city)
```

```python
#30 - Count Phrase Matches
#"Python is used in deep learning"
#"deep learning", "python"

import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")

text = "Python is used in deep learning"
doc = nlp(text)

matcher = PhraseMatcher(nlp.vocab)

phrases = ["deep learning", "python"]
patterns = [nlp(phrase) for phrase in phrases]

matcher.add("TECH_TERMS", patterns)

matches = matcher(doc)

count = len(matches)

print(count)
```
