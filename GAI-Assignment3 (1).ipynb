```python
#2 - For every token in the third sentence, print the token text, the POS tag,
# the fine-grained TAG tag, and the description of the fine-grained tag.

import spacy

from spacy import explain

nlp = spacy.load("en_core_web_sm")

with open("peterrabbit.txt", "r", encoding="utf-8") as file:
    text = file.read()

doc = nlp(text)

sentences = list(doc.sents)

third_sentence = sentences[2]

for token in third_sentence:
    print(
        token.text,
        token.pos_,
        token.tag_,
        explain(token.tag_)
    )
```

```python
#3 - Provide a frequency list of POS tags from the entire document

import spacy

from spacy import explain

nlp = spacy.load("en_core_web_sm")

with open("peterrabbit.txt", "r", encoding = "utf-8") as file:
    text = file.read()

doc = nlp(text)

pos_counts = doc.count_by(spacy.attrs.POS)

for k, v in sorted(pos_counts.items()):
    print(f"{k}. {doc.vocab[k].text} ({explain(doc.vocab[k].text)}): {v}")
```

```python
#4 - CHALLENGE: What percentage of tokens are nouns?

import spacy

from spacy import explain

nlp = spacy.load("en_core_web_sm")

with open("peterrabbit.txt", "r", encoding="utf-8") as file:
    text = file.read()

doc = nlp(text)

noun_count = 0

for token in doc:
    if token.pos_ == "NOUN":
        noun_count += 1

total_tokens = len(doc)

percentage = (noun_count / total_tokens) * 100

print("Number of nouns:", noun_count)
print("Total tokens:", total_tokens)
print("Percentage of nouns: {:.2f}%".format(percentage))


```

```python
#8 - CHALLENGE: How many sentences contain named entities?

import spacy

nlp = spacy.load("en_core_web_sm")

with open("peterrabbit.txt", "r", encoding="utf-8") as file:
    text = file.read()

doc = nlp(text)

count = 0

for sent in doc.sents:
    if len(sent.ents) > 0:
        count += 1

print("Sentences containing named entities:", count)
```
